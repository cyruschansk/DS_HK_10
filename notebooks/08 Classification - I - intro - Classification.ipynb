# Classification

Classification is a type of supervised learning where the goal is to predict class labels of instances based on past observations. The class labels are discrete, unordered values that can be considered as group memberships to which instances belong.

Many classification problems are *binary* problems. For example, if you have a set of emails and you want to classify them as spam or not. However, the set of labels does not have to be binary. A predictive model can assign any class label that was presented in the training dataset to a new, unlabeled instance. A typical example of a *multi-class* classification problem is the task of handwritten character recognition.

It may not be immediately obvious, but many regression problems can also be redefined as classification problems. Consider a situation where you have a dataset containing people's yearly salaries. How would you transform this dataset in order to identify people that are more likely to be upper-class, middle-class, or lower-class?
